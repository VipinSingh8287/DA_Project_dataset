# Hypothesis: There exists a positive correlation between, 
# "Renewable energy share in the total final energy consumption (%)" and "Low-carbon electricty (%)"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('global-data-on-sustainable-energy.csv')

df_subset = df[['Entity', 'Renewable energy share in the total final energy consumption (%)', 'Low-carbon electricity (% electricity)']]

df_subset = df_subset.dropna(subset=['Renewable energy share in the total final energy consumption (%)', 'Low-carbon electricity (% electricity)'])

df_subset = df_subset[df_subset['Low-carbon electricity (% electricity)'] != 0]
```

```python
correlation, _ = stats.spearmanr(df_subset['Renewable energy share in the total final energy consumption (%)'], df_subset['Low-carbon electricity (% electricity)'])
print(f"Correlation coefficient: {correlation:.4f}")
```

```python
alpha = 0.05
n = len(df_subset)
r = correlation
t_score = r * np.sqrt((n - 2)/(1 - r**2))
p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=n-2))
```

```python
print(f"Spearman correlation coefficient: {r:.4f}")
print(f"T-score: {t_score:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
x = df_subset['Renewable energy share in the total final energy consumption (%)']
y = df_subset['Low-carbon electricity (% electricity)']
plt.scatter(x, y, s=1, c='orange')

plt.xlabel('Renewable energy share in the total final energy consumption (%)')
plt.ylabel('Low-carbon electricity (% electricity)')
plt.title('Scatter Plot: Renewable Energy(%) vs Low-Carbon Electricity(%)')

plt.show()
```
