```python
import pandas as pd
```

```python
df =pd.read_excel("./JAN2020-Renewable energy Installed and Potential Capacity.xlsx")
```

```python
df = df.dropna()
```

```python
df
```

```python
df1 = pd.DataFrame(columns=["Date","Source", "State","Potential Capacity","Installed Capacity"])
```

```python
df1
```

```python
for i in range(0,380,2):
    row1 = df.iloc[i]
    row2 = df.iloc[i+1]
    source = row1.iloc[0]
    state = row1.iloc[1]
    date = row1.iloc[4]
    potential = row1.iloc[3]
    installed = row2.iloc[3]
    new_row = {
        'Date':date,
        'Source':source,
        'State':state,
        'Potential Capacity':potential,
        'Installed Capacity':installed
    }
    df1.loc[len(df1.index)] = new_row
```
